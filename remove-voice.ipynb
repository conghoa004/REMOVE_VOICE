# Vocal Separation using BS-Roformer

This notebook demonstrates how to separate vocals and instrumentals from an audio file using the state-of-the-art **BS-Roformer** model.

## Step 1: Install BS-Roformer Library

Install the `bs-roformer-infer` package which provides the command-line tools for model downloading and inference.

```python
# Install the BS-Roformer inference library silently
!pip install -q bs-roformer-infer
```

## Step 2: List Available Models

Check the list of available pre-trained models to choose the most suitable one.

```python
# Display the list of available pre-trained BS-Roformer models
!bs-roformer-download --list-models
```

## Step 3: Download the Vocals Resurrection Model

Download the recommended high-performance vocals resurrection model by Unwa. The downloaded assets will be stored in the `models` directory.

```python
# Download the high-quality vocals separation model by Unwa
!bs-roformer-download \
--model roformer-model-bs-roformer-vocals-resurrection-by-unwa \
--output-dir models
```

## Step 4: Install FFmpeg

FFmpeg is required for audio decoding, encoding, and conversion.

```python
# Install FFmpeg package silently using apt-get
!apt-get -qq install ffmpeg
```

## Step 5: Download Audio File from Google Drive

Download your custom audio file from a shared Google Drive link using the `gdown` Python library.

```python
import gdown

# Provide the sharing link of your Google Drive file
link_drive = "https://drive.google.com/file/d/<id>/view?usp=drive_link"

# Extract the file ID from the sharing URL
file_id = link_drive.split("/")[5]
url = f"https://drive.google.com/uc?id={file_id}"

# Download the file to the current workspace
gdown.download(url, "input_music.mp3", quiet=False)
```

## Step 6: Convert Input Audio to WAV Format

Convert the input audio file to WAV format with a 44100Hz sample rate and stereo (2 channels) layout for optimal compatibility.

```python
# Convert input MP3 audio to a 44100Hz, stereo WAV file
!ffmpeg -i "input_music.mp3" \
-ar 44100 -ac 2 \
"output_music.wav"
```

## Step 7: Prepare Input Directory

Create a dedicated folder for inputs and copy the target WAV file there.

```python
# Create the input directory and copy the converted WAV file into it
!mkdir -p /kaggle/working/input
!cp "output_music.wav" /kaggle/working/input/
```

## Step 8: Run BS-Roformer Inference

Run the source separation inference to separate the vocal and instrumental components of the track.

```python
# Run BS-Roformer inference to separate vocals and instrumentals
!bs-roformer-infer \
--config_path /kaggle/working/models/roformer-model-bs-roformer-vocals-resurrection-by-unwa/config_bs_roformer_vocals_resurrection_unwa.yaml \
--model_path /kaggle/working/models/roformer-model-bs-roformer-vocals-resurrection-by-unwa/bs_roformer_vocals_resurrection_unwa.ckpt \
--input_folder "/kaggle/working/input" \
--store_dir "/kaggle/working/output"
```
